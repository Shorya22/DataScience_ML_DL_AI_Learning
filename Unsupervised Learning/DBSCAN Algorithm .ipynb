```python
# Import Libs
import pyforest  # For Basic Libs pandas,numpy,matplorlib,seaborn etc.
from sklearn.cluster import DBSCAN # DBSCAN Algorithm
from sklearn.datasets import make_moons # To create Non Linear Dataset
from sklearn.model_selection import train_test_split #TTS
```

```python
# Create non linear dataset with some noise
X,y= make_moons(n_samples= 2200,noise=0.1,random_state=42)
```

```python
# Scatter Plot For Dataset
plt.scatter(X[:,0],X[:,1])
plt.show()
```

```python
# Splitting Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
# Implement DBSCAN Cluster Algorithm and Model Training
cluster= DBSCAN(eps=0.13,min_samples=7,n_jobs=-1)
cluster.fit_predict(X_train)
y_train_labels= cluster.labels_
```

```python
# Test Model and Labels
cluster.fit_predict(X_test)
y_test_labels= cluster.fit_predict(X_test)
```

```python
# Scatter Plot For Training Data After Clustering (X_train) & Finding Outliers
plt.scatter(X_train[:,0],X_train[:,1],c= y_train_labels)
plt.show()
```

```python
# Scatter Plot For Testing Data After Clustering (X_test) & Finding Outliers
plt.scatter(X_test[:,0],X_test[:,1],c= y_test_labels )
plt.show()
```
