# Linear Regression Mathematics:


```python
import numpy as np
```

```python
# Height (in inches)
X = np.array([65, 68, 72, 63, 70, 66, 69, 67, 64, 71])

# Weight (in KG)
y = np.array([150, 160, 180, 140, 175, 155, 165, 162, 145, 170])
```

```python
#Steps To Perform:

# Step_1: Define Some Terms which will use in next steps and formula
        # Define b which is intercept and initial value b=0
        # Define m which is slope or coefficient and initial value m=0
        # Define len of data which is n= len(X)
        # Define epoches or iterations epoch= 100,500,1000.....
        # Define Learning Rate which is alpha= 0.1,0.01,0.001......
        
# Step_2: Now Perform iteration steps include:
        # Find y_pred  (z=m.x+b)
        # Find Cost Function cost= 1/2n.np.sum((y_pred-y_true)**2)
        # Find Gradient Descent For m and b
        # then change in m and b with alpha so cost function and loss reduce and reach minima
```

```python
# Step_1:
b=0
m=0
n= len(X)
epoches=10000000
alpha= 0.0002
```

```python
#Step_2:
for epoch in range (epoches):
    
    y_pred= m*X+b
    
    loss= (1/2*n)*np.sum((y_pred-y)**2)
    
    gradient_m= (1/n)*np.sum((y_pred-y)*X)
    gradient_b= (1/n)*np.sum(y_pred-y)
    
    m-= alpha*gradient_m
    b-= alpha*gradient_b
    
    if epoch % 100000 == 0:
        print(f"epoches {epoch+1},loss:{loss},m: {m},b: {b}")
    
    
    
    
```
