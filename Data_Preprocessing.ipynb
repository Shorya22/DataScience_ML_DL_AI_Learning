**Data Preprocessing (Data Cleaning, Feature Scaling, One-Hot Encoding, Handling Missing Values):**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
data= pd.read_csv(r'F:\IT Learning\MY PROJECTS\Python Projects\Data Cleaning Datasets\titanic_data.csv')
```

```python
data
```

```python
data.isna().sum()
```

```python
imputer=SimpleImputer(strategy= 'mean')
```

```python
imputed_age= imputer.fit_transform(data[['age']])
```

```python
data['age']= imputed_age.astype('int')
```

```python
data
```

```python
x= data[['sex']]
y= data[['adult_male']]
```

```python
ohe= OneHotEncoder(sparse_output= False)
```

```python
ohe_sex= ohe.fit_transform(x[['sex']]).astype('int')
```

```python

features= ohe.get_feature_names_out(['sex'])
x= pd.DataFrame(ohe_sex, columns= features)
```

```python
x
```

```python
ohe_n= OneHotEncoder(sparse_output= False)
ohe_adult_male = ohe_n.fit_transform(y[['adult_male']]).astype('int')
features = ohe_n.get_feature_names_out(['adult_male'])
y = pd.DataFrame(ohe_adult_male, columns=features)

```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2)
```
